```python
import spacy
nlp = spacy.load("en_core_web_sm")
doc = nlp("I have a severe headache.")
for token in doc:
    print(token.text, "-->", token.pos_) # Prints: headache -> NOUN
```

```python
import spacy
from spacy.matcher import Matcher


nlp = spacy.load("en_core_web_sm")
matcher = Matcher(nlp.vocab)
pattern = [
    {"POS": "ADJ"},   
    {"POS": "NOUN"}   
]


matcher.add("ADJECTIVE_NOUN_PATTERN", [pattern])
doc = nlp("I am in a yellow car with severe headache and mild nausea.")
matches = matcher(doc)
print("Found patterns:")
for match_id, start, end in matches:

    span = doc[start:end]
    print(span.text)
```
